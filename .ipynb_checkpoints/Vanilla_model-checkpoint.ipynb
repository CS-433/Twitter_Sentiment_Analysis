```python
with open('vocab.pkl', 'rb') as f:
    vocab = pickle.load(f)
vocab_size = len(vocab)
```

```python
def find_token_idx(tkn, vocab):
    if tkn in vocab:
        return vocab[tkn]
    else:
        return -1
```

```python
def encode_token(tkn, u, vocab):
    idx = find_token_idx(tkn, vocab)
    if idx >= 0:
        return u[idx, :]
    else:
        return np.zeros(u.shape[1])
```

```python
def encode_tweet(tweet, u, vocab):
    # Tokenize tweet
    tw = tweet.split(' ')
    
    return np.concatenate([encode_token(w, u=u, vocab=vocab) for w in tw]).reshape(-1, u.shape[1]).mean(axis=0)
```

```python
def load_dataset(filename, label, embeddings, vocab):
    tweets = []
    count = 0
    with open(filename, 'r', encoding='utf-8') as f:
        for tweet in f:
            tweets.append(encode_tweet(tweet, embeddings, vocab))
            count += 1
            
    tX = np.concatenate(tweets).reshape(-1, embeddings.shape[1])
    
    # Assign labels
    y = np.repeat(label, count)
    
    return tX, y
```

```python
with open('neg_train.txt', 'r') as neg:
    ln = neg.readline().split(' ')
    print(ln)
```

```python
pos_dataset, pos_labels = load_dataset('pos_train.txt', 1)
neg_dataset, neg_labels = load_dataset('neg_train.txt', -1)
```

```python
tX = np.concatenate([pos_dataset, neg_dataset], axis=0)
y = np.concatenate([pos_labels, neg_labels])
tX, y
```

```python
# Shuffle train data
random_shuffle = np.random.permutation(y.shape[0])
tX = tX[random_shuffle]
y = y[random_shuffle]
```

```python
# Normalize data
tX = tX - tX.mean(axis=0)
tX = tX / tX.std(axis=0)
```

```python
y
```

```python
log_reg = LogisticRegression(solver='saga').fit(tX, y)
```

```python
lin_reg = LinearRegression(normalize=True)
lin_reg.fit(tX, y)
```

```python
lin_reg.score(tX, y)
```

```python
log_reg.score(tX, y)
```

## GloVe

```python
glove_emb = np.load('embeddings_full.npy')
```

```python
glove_emb.shape
```

```python
def load_vocab(filename):
    with open(filename, 'rb') as f:
        vocab = pickle.load(f, encoding='utf-8')
    vocab_size = len(vocab)
    print("Loaded vocab of size: %d" % vocab_size)
    return vocab
```

```python
vocab_full = load_vocab('vocab_full.pkl')
```

```python
pos_full_data, pos_full_labels = load_dataset('twitter-datasets/twitter-datasets/train_pos_full.txt', label=1, embeddings=glove_emb, vocab=vocab_full)
```

```python
neg_full_data, neg_full_labels = load_dataset('twitter-datasets/twitter-datasets/train_neg_full.txt', label=-1, embeddings=glove_emb, vocab=vocab_full)
```

```python
tX_full = np.concatenate([pos_full_data, neg_full_data], axis=0)
y_full = np.concatenate([pos_full_labels, neg_full_labels])

# Sanity check
assert tX_full.shape[0] == y_full.shape[0]

tX_full, y_full
```

```python
def shuffle_data(tX, y):
    random_shuffle = np.random.permutation(y.shape[0])
    return tX[random_shuffle], y[random_shuffle]
```

```python
tX_full, y_full = shuffle_data(tX_full, y_full)
```

```python
log_reg = LogisticRegression(solver='saga')
```

```python
log_reg.fit(tX_full, y_full)
```

```python
log_reg.score(tX_full, y_full)
```
